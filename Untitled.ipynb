# Adding colums High_Low_Speed and High_Low_HP

```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
data.describe()
```

```python
data["Speed"].mean()
```

```python
Speed_mean = data["Speed"].mean()
def set_speed(val):
    if val > Speed_mean:
        return "Speed high"
    else:
        return "Speed low"
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```

```python
data["HP"].mean()
```

```python
HP_mean = data["HP"].mean()
def set_HP(val):
    if val > HP_mean:
        return "HP high"
    else:
        return "HP low"

data["HP_high_low"] = data["HP"].apply(set_HP)
data
```
